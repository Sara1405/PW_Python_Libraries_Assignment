```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
from os.path import samefile
# Loading Dataset
sales_df = pd.read_excel('/content/retail_dataset .xlsx')
print(sales_df.head)
```

```python
# Q1. Load the dataset and display first 5 rows along with dataset information
# display first 5 rows
print(sales_df.head())

# display dataset information
print(sales_df.info())
```

```python
# Q2. Identify missing values in each column
# checking missing values
missing_value = sales_df.isnull().sum()
print(missing_value)

# Total missing values
print("Total_missing_values:",missing_value.sum())
```

```python
# Q3. Replace missing values in Quantity and Price columns with mean values
# Fill missing Quantity values with mean
sales_df['Quantity'] = sales_df['Quantity'].fillna(sales_df['Quantity'].mean())

# Fill missing Price values with mean
sales_df['Price'] = sales_df['Price'].fillna(sales_df['Price'].mean())

print(sales_df)
```

```python
# Q4. Remove rows where Product Category or Region contains missing values
# Remove rows with missing Product Category or Region
sales_df = sales_df.dropna(subset=['Product Category', 'Region'])

print(sales_df)
```

```python
# Q5. Create a new Revenue column
# Create Revenue column
sales_df['Revenue'] = np.multiply(sales_df['Quantity'], sales_df['Price'])

# Display updated dataset
print(sales_df.head())
```

```python
# Q6. Calculate total revenue using NumPy
# Total revenue
total_revenue = np.sum(sales_df['Revenue'])

print("Total Revenue:", total_revenue)
```

```python
# Q7. Group dataset by Product Category and calculate total revenue
# Revenue by Product Category
category_revenue = sales_df.groupby('Product Category')['Revenue'].sum()

print(category_revenue)
```

```python
# Q8. Identify Top 3 and Bottom 3 Product Categories by Revenue
# Sort category revenue
sorted_category_revenue = category_revenue.sort_values(ascending=False)

# Top 3 categories
print("Top 3 Categories")
print(sorted_category_revenue.head(3))

# Bottom 3 categories
print("Bottom 3 Categories")
print(sorted_category_revenue.tail(3))
```

```python
# Q9. Group dataset by Region and calculate total revenue
# Revenue by Region
region_revenue = sales_df.groupby('Region')['Revenue'].sum()

print(region_revenue)

# Highest revenue region
highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print("Highest Revenue Region:", highest_region)
print("Lowest Revenue Region:", lowest_region)
```

```python
# Q10. Extract Month from Date column and calculate monthly revenue
# Convert Date column into datetime format
sales_df['Date'] = pd.to_datetime(sales_df['Date'])

# Extract month name
sales_df['Month'] = sales_df['Date'].dt.month_name()

# Monthly revenue
monthly_revenue = sales_df.groupby('Month')['Revenue'].sum()

print(monthly_revenue)

# Mean, Median and Standard Deviation of Revenue
mean_revenue = np.mean(sales_df['Revenue'])
median_revenue = np.median(sales_df['Revenue'])
std_revenue = np.std(sales_df['Revenue'])

print("Mean Revenue:", mean_revenue)
print("Median Revenue:", median_revenue)
print("Standard Deviation:", std_revenue)
```
